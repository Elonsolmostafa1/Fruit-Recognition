```python
from numba import jit,njit
from time import time

# helper function
@njit()
def extended_gcd(a,b):
    # assume a>=b
    if b==0 :
        d,x,y=a,1,0
    else:
        d,p,q=extended_gcd(b,a%b)
        x=q
        y=p-q*(a//b)
    return d,x,y


@jit(nopython=True)
def right_to_left(b,e,n):
    a,s=1,b%n
    while e>0:
        if e%2==1:
            a=(a*s)%n
        e//=2
        s=(s*s)%n
    return a

@jit(nopython=True)
def left_to_right(b,e,n):
    bits=[]
    while e:
        bits.insert(0,e%2)
        e//=2
    s=b
    for bit in bits[1:]:
        s=(s*s)%n
        if bit==1:
            s*=b
    return s


@jit(nopython=True)
def crt(b,e,p,q):
    yp=right_to_left(b,e,p)
    yq=right_to_left(b,e,q)
    if p>q:
        d,xp,xq=extended_gcd(p,q)
    else:
        d,xq,xp=extended_gcd(q,p)
        
    return (yp*q*xq + yq*p*xp)%(p*q)

b,e,n,p,q=11,12354,40806607,5153,7919

# Run one time to compile
right_to_left(b,e,n)
left_to_right(b,e,n)
crt(b,e,p,q)

# run again to calculate the time
start=time()
result=right_to_left(b,e,n)
dur=time()-start
print("right to left took {}s,the result is {}".format(dur,result))

start=time()
result=left_to_right(b,e,n)
dur=time()-start
print("left to right took {}s,the result is {}".format(dur,result))

start=time()
result=crt(b,e,p,q)
dur=time()-start
print("crt took {}s,the result is {}".format(dur,result))
```
